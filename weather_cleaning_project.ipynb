```python
## QUESTION:
# Step-2 Finding a summer break
# According to meteorologists, summer extends for the whole months of June, July, and August 
# in the northern hemisphere and the whole months of December, January,
# and February in the southern hemisphere. So create a dataframe that 
# holds just those months using the `datetime` index.
```

```python
#Created a variable 'summer_df' in the northern hemisphere
#Turned the var into a dataframe
northern_summer = df[df['datetime'].dt.month.isin([6,7,8])]
print(".... NORTHERN SUMMER NOTEBOOK ....")
print(f"Summer days collected: {len(northern_summer)}")
print(f"Summer date range: {northern_summer['datetime'].min()} to {northern_summer['datetime'].max()}")
```

```python
#function to get the season any inputted month is
def get_season(month):
    if month in [12,1,2]:
        return'winter'
    elif month in [3,4,5]:
        return 'spring'
    elif month in [6,7,8]:
        return 'summer'
    else:
        return 'fall'
df['season'] = df['datetime'].dt.month.apply(get_season)    
```
